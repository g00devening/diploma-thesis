```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
n = 3000  # Размер матрицы
X = np.random.randn(n, n)
A = X / np.sqrt(n)
```

```python
eigenvalues = np.linalg.eigvals(A)

real_part = eigenvalues.real
imag_part = eigenvalues.imag
```

```python
plt.figure(figsize=(8, 8))

# собственные значения
plt.scatter(real_part, imag_part, s=3, alpha=0.6, color='lime', label='Собственные числа')

# единичная окружность
theta = np.linspace(0, 2*np.pi, 200)
plt.plot(np.cos(theta), np.sin(theta), 'g-', linewidth=2, label='Единичный круг')

plt.axhline(y=0, color='black', linestyle='-', alpha=0.3, linewidth=0.5)
plt.axvline(x=0, color='black', linestyle='-', alpha=0.3, linewidth=0.5)
plt.xlabel(r'Re($\lambda$)')
plt.ylabel(r'Im($\lambda$)')
plt.title(f'Круговой закон для случайной матрицы размера {n}x{n}')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend()
#plt.savefig('./pictures for diplom/circular_law.png')
plt.show()
```
